```python
def save_frame(frame):
    now=datetime.datetime.now()
    filename = "frames/frame_{}.png".format(now.strftime("%Y-%m-%d_%H-%M-%S"))
    cv2.imwrite(filename, frame)
```

```python
while True:
    check, frame = cam.read()#capture frame from video
    cv2.imshow('video', frame)
    key=cv2.waitKey(1)
    if key == ord('s'):
        save_frame(frame)
    if key==27 :
        break
cam.release()
cv2.destroyAllWindows()
```

```python
#get images
path_to_image=("/home/agj/VSCODE/Jeffrin/Project obj-measur-sys/frames/")
def get_latest_image(directory, valid_extensions=('jpg', 'jpeg', 'png')):
    #get filepaths of all files and dirs in the given dir
    valid_files=[os.path.join(directory, filename) for filename in os.listdir(directory)]
    valid_files=[f for f in valid_files if '.' in f and f.rsplit('.',1)[1] in valid_extensions and os.path.isfile(f)]
    #f.rsplit('.',1)[1] split the string once and select the string after ".".
    #os.path.isfile(f) checks the string f corresponds to a valid file or not.
    if not valid_files:
        raise ValueError("No valid images in %s" % directory)
    return max(valid_files, key=os.path.getmtime)
last_image=get_latest_image(path_to_image)
```

```python
#read image
imge=cv2.imread(last_image)
imge_cpy=imge.copy()

cv2.imshow('image', imge)
cv2.imshow('Img_cpy', imge_cpy)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
gray=cv2.cvtColor(imge, cv2.COLOR_BGR2GRAY)
cv2.imshow('gray', gray)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
blur=cv2.GaussianBlur(gray, (3,3), 0)
cv2.imshow("blur", blur)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
edges=cv2.Canny(gray, 250,100) #default sobelkernel=3
edge_cpy=cv2.Canny(blur, 180,210) #default sobelkernel=3
contours, _= cv2.findContours(edge_cpy, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cv2.imshow("canny_gray", edges)
cv2.imshow("canny_1", edge_cpy)
cv2.waitKey(50000)
cv2.destroyAllWindows()
print(len(contours))
```

```python
#removing small contours
_, thresh=cv2.threshold(edge_cpy, 127, 255, cv2.THRESH_BINARY)
# Loop through contours
for cnt in contours:
    if cv2.contourArea(cnt)< 5 :
        cv2.drawContours(thresh, [cnt], 0, 0, -1)
    

cv2.imshow('Out_put Image', thresh)
cv2.imshow("rec", imge)

cv2.waitKey(50000)
cv2.destroyAllWindows()
```

```python
# Find the contour with the largest area, which is most likely to be the rectangle
max_area = 0
max_contour = None
for contour in contours:
    area = cv2.contourArea(contour)
    
    print(area)
    if area > max_area:
        max_area = area
        max_contour = contour

# Get the coordinates and dimensions of the rectangle bounding the contour
x, y, w, h = cv2.boundingRect(max_contour)
topleft_axes=(x,y)
bottomright_axes=(x+w,y+h)
width, height=w, h
w
```

```python
points=cv2.findNonZero(thresh)
rect=cv2.minAreaRect(points)#minimum area rotated rectangular box around contours
box=cv2.boxPoints(rect)#obtaining 4 cornerpoints of rotated rectangle as [p1,p2,p3,p4]
box=np.int0(box)
print(box)
print("cnt is: ", cnt)
cv2.drawContours(imge, [box], 0, (0,0,255),2) #3-> index of contour
cv2.imshow("rec", imge)
cv2.waitKey(50000)
cv2.destroyAllWindows()
```

```python
cv2.imshow('edge_cpy', edge_cpy)
cv2.imshow('edges', edges)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
contours15, hierarchy = cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

max_contour_area = 0
largest_contour_index = -1
for i, contour in enumerate(contours15):
    contour_area = cv2.contourArea(contour)
    if contour_area > max_contour_area:
        max_contour_area = contour_area
        largest_contour_index = i
        
# Draw a white filled polygon of the largest contour
mask = np.zeros(edges.shape, np.uint8)
cv2.drawContours(mask, [contours15[largest_contour_index]], 0, 255, -1)

# Erase all the contours15 outside the largest contour area
erased = cv2.bitwise_and(edges, mask)

# Display the resulting image
cv2.imshow('Edges with largest contour', erased)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

contours1, hierarchy = cv2.findContours(erased, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours1))
# Find the index of the largest contour based on area
max_contour_idx = max(range(len(contours1)), key=lambda i: cv2.contourArea(contours1[i]))

# Create a mask image of the same size as the original image
mask = np.zeros_like(erased)

# Draw the largest contour in white on the mask image
cv2.drawContours(mask, contours1, max_contour_idx, (255, 255, 255), cv2.LINE_AA)

# Invert the mask image
mask_inv = cv2.bitwise_not(mask)

# Apply the mask to the Canny image to remove the largest contour
edges_without_largest_contour = cv2.bitwise_and(erased, mask_inv)

# Display the result
cv2.imshow('Edges without largest contour', edges_without_largest_contour)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Remove small contours
min_contour_area = 100  # set the minimum contour area threshold
for contour in contours:
    contour_area = cv2.contourArea(contour)
    if contour_area < min_contour_area:
        cv2.drawContours(edges_without_largest_contour, [contour], 0, 0, -1)

# Display the resulting image
cv2.imshow('Edges with small contours removed', edges_without_largest_contour)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

#removing small contours
_, thresh3=cv2.threshold(edges_without_largest_contour, 127, 255, cv2.THRESH_BINARY)
contours3, _= cv2.findContours(thresh3, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Loop through contours
for cnt in contours3:
    if cv2.contourArea(cnt)< 10 :
        cv2.drawContours(thresh3, [cnt], 0, 0, -1)
    

cv2.imshow('thresh3', thresh3)
cv2.imshow("remo_edges", edges_without_largest_contour)

cv2.waitKey(50000)
cv2.destroyAllWindows()
```
